```python
df_services.columns = df_services.columns.str.strip()
df_info.columns = df_info.columns.str.strip()

df_services.dropna(subset=["Rndrng_Prvdr_Org_Name", "DRG_Desc"], inplace=True)
df_info.dropna(subset=["Hospital Name", "ZIP Code"], inplace=True)
```

```python
df_services["Rndrng_Prvdr_Org_Name"] = df_services["Rndrng_Prvdr_Org_Name"].str.upper().str.strip()
df_info["Hospital Name"] = df_info["Hospital Name"].str.upper().str.strip()

df_services["Rndrng_Prvdr_Zip5"] = df_services["Rndrng_Prvdr_Zip5"].astype(str).str[:5]
df_info["ZIP Code"] = df_info["ZIP Code"].astype(str).str[:5]
```

```python
df_merged = pd.merge(df_services, df_info, left_on=["Rndrng_Prvdr_Org_Name", "Rndrng_Prvdr_Zip5"], right_on=["Hospital Name", "ZIP Code"], how="inner")
```

```python
df_model = df_merged[[
    "Rndrng_Prvdr_Org_Name", "City", "State", "DRG_Desc",
    "Avg_Submtd_Cvrd_Chrg", "Avg_Tot_Pymt_Amt", "Avg_Mdcr_Pymt_Amt",
    "Hospital overall rating", "Emergency Services", "Hospital Type"
]].copy()
```

```python
from sklearn.preprocessing import LabelEncoder

le_drg = LabelEncoder()
df_model["DRG_Cd"] = le_drg.fit_transform(df_model["DRG_Desc"])

le_state = LabelEncoder()
df_model["Rndrng_Prvdr_State_FIPS"] = le_state.fit_transform(df_model["State"])
```

```python
for col in ["Avg_Submtd_Cvrd_Chrg", "Avg_Tot_Pymt_Amt", "Avg_Mdcr_Pymt_Amt"]:
   df_model[col] = df_model[col].replace('[$,]', '', regex=True).astype(float)

```

```python
features = df_model[[
    "DRG_Cd", "Rndrng_Prvdr_State_FIPS", "Avg_Submtd_Cvrd_Chrg", "Avg_Mdcr_Pymt_Amt"
]]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors=5, metric='euclidean')
knn.fit(features_scaled)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
def recommend_hospitals(emergency_services, state, max_cost=None):
    service_code = le_drg.transform([str(emergency_services)])[0]
    state_code = le_state.transform([str(state)])[0]

    # Filter the sample based on the service and state
    sample = df_model[
        (df_model["DRG_Cd"] == service_code) &
        (df_model["Rndrng_Prvdr_State_FIPS"] == state_code)
    ]

    # If max_cost is provided, filter based on it
    if max_cost:
        sample = sample[sample["Avg_Submtd_Cvrd_Chrg"].notna()]
        sample = sample[sample["Avg_Submtd_Cvrd_Chrg"] <= max_cost]

    # Apply scaling to the selected features
    sample_features = scaler.transform(sample[[
        "DRG_Cd", "Rndrng_Prvdr_State_FIPS", "Avg_Submtd_Cvrd_Chrg", "Avg_Mdcr_Pymt_Amt"
    ]])

    # Find nearest neighbors
    distances, indices = knn.kneighbors(sample_features)

    # Return the recommended hospitals
    return sample.iloc[indices[0]]
```

```python
def recommend_hospitals(emergency_services, state, max_cost=None, top_n=5):
    try:
        service_code = le_drg.transform([emergency_services])[0]
        state_code = le_state.transform([state])[0]
    except:
        print("Invalid service or state name.")
        return pd.DataFrame()

    # Filter data for matching service and state
    filtered_df = df_model[
        (df_model["DRG_Cd"] == service_code) &
        (df_model["Rndrng_Prvdr_State_FIPS"] == state_code)
    ]

    if max_cost:
        filtered_df = filtered_df[filtered_df["Avg_Submtd_Cvrd_Chrg"] <= max_cost]

    if filtered_df.empty:
        print("No matching hospitals found.")
        return pd.DataFrame()

    # Prepare features for KNN
    X = filtered_df[["DRG_Cd", "Rndrng_Prvdr_State_FIPS", "Avg_Submtd_Cvrd_Chrg", "Avg_Mdcr_Pymt_Amt"]]
    X_scaled = scaler.transform(X)

    # Find nearest neighbors
    distances, indices = knn.kneighbors(X_scaled[:1], n_neighbors=min(top_n, len(X_scaled)))

    recommended = filtered_df.iloc[indices[0]]
    return recommended[["Rndrng_Prvdr_Org_Name", "City", "Avg_Submtd_Cvrd_Chrg", "Avg_Mdcr_Pymt_Am", "Hospital overall rating"]]
```

```python
df_model["Cost_Bucket"] = pd.cut(df_model["Avg_Submtd_Cvrd_Chrg"], bins=5, labels=False)
states = df_model[["DRG_Cd", "Rndrng_Prvdr_State_FIPS", "Cost_Bucket"]].dropna().drop_duplicates().values.tolist()
actions = list(range(len(df_model)))
```

```python
def get_reward(hospital_row):
    medicare_payment = hospital_row["Avg_Mdcr_Pymt_Amt"]
    rating = hospital_row.get("Hospital overall rating", 3)
    cost = hospital_row["Avg_Submtd_Cvrd_Chrg"]

    # Reward high Medicare coverage, good rating, low cost
    return (medicare_payment / 10000) + (int(rating) * 0.5) - (cost / 50000)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

class QNetwork(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(QNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.fc2 = nn.Linear(128, 64)
        self.out = nn.Linear(64, output_dim)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.out(x)
```

```python
def get_reward(hospital_row):
    medicare_payment = hospital_row["Avg_Tot_Pymt_Amt"]
    rating_str = hospital_row["Hospital overall rating"]
    cost = hospital_row["Avg_Submtd_Cvrd_Chrg"]
    
    try:
        rating = int(rating_str)
    except (ValueError, TypeError):
        rating = 0
    return (medicare_payment / 10000) + (rating * 0.5) - (cost / 50000)
```

```python
# Prepare inputs
state_input = torch.tensor(states, dtype=torch.float32)
num_states = state_input.shape[1]
num_actions = len(actions)

# Init model
q_net = QNetwork(num_states, num_actions)
optimizer = optim.Adam(q_net.parameters(), lr=0.001)
gamma = 0.9  # Discount factor

# Training loop
for episode in range(1000):
    state_idx = random.choice(range(len(states)))
    state = torch.tensor(states[state_idx], dtype=torch.float32).unsqueeze(0)
    
    # Epsilon-greedy action
    if random.random() < 0.1:
        action_idx = random.choice(actions)
    else:
        q_values = q_net(state)
        action_idx = torch.argmax(q_values).item()
    
    hospital = df_model.iloc[action_idx]
    reward = get_reward(hospital)

    # Assume next_state same as current for now
    next_state = state
    target = reward + gamma * torch.max(q_net(next_state)).item()

    # Loss & update
    q_val = q_net(state)[0][action_idx]
    loss = F.mse_loss(q_val, torch.tensor(target, dtype=torch.float32))


    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if episode % 100 == 0:
        print(f"Episode {episode}, Reward: {reward:.2f}, Loss: {loss.item():.4f}")
```

```python
def rl_recommendation(service_name, state_name, cost_value):
    service_code = le_drg.transform([service_name])[0]
    state_code = le_state.transform([state_name])[0]
    cost_bucket = int(pd.cut([cost_value], bins=5, labels=False)[0])

    state = torch.tensor([[service_code, state_code, cost_bucket]], dtype=torch.float32)
    q_values = q_net(state)
    top_indices = torch.topk(q_values, k=5).indices.squeeze().tolist()

    return df_model.iloc[top_indices][[
        "Provider Name", "City_x", "Average Covered Charges",
        "Average Medicare Payments", "Hospital overall rating"
    ]]
```
