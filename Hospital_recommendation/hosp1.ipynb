```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# Attempt to read the CSV file with a different encoding (e.g., 'ISO-8859-1', 'latin1', or 'cp1252')
try:
    df = pd.read_csv('Hospital General Information.csv', encoding='ISO-8859-1')  # or try 'latin1' or 'cp1252'
except FileNotFoundError:
    print("Error: File 'Hospital General Information.csv' not found.")
    exit()

col_names = df.columns.to_list()

for col in col_names:
    print(col)
```

```python
df = df[["Provider ID","Hospital Name","Address","City","State","ZIP Code","County Name","Phone Number","Hospital Type","Hospital Ownership","Emergency Services","Hospital overall rating","Mortality national comparison","Patient experience national comparison","Effectiveness of care national comparison"]]
```

```python
#Handling missing value
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
df['County Name'] = imputer.fit_transform(df[['County Name']]).ravel()#.ravel() flattens the result into 1D
print(df['County Name'].head())
```
